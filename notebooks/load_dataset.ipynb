```python
def loadAde20kImage(file):
    file_seg = file.replace(".jpg","_seg.png")
    with Image.open(file_seg) as io:
        seg = np.array(io)
    
    R = seg[:,:,0]
    G = seg[:,:,1]
    ObjectClassMasks = (R / 10).astype(np.int32) * 256 + (G.astype(np.int32))
    return {
        "img_name":file,
        "segm_name":file_seg,
        "class_mask":ObjectClassMasks
    }
```
